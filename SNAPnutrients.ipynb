```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import sqlite3 as sql

FNDDS = pd.read_csv("FNDDS_Nutrition2.csv")
WIC1 = pd.read_csv("WIC+portionsdraft.csv")
WICcosts = pd.read_csv("WIC+cost2.csv")
```

```python
nutrition1 = pd.merge(WIC1,FNDDS,on="Food Code", how = "inner", )[["Food and Product Sizes", "# 100 grams", "Protein (g)", "Carbohydrate (g)", "Fiber, total dietary (g)", "Total Fat (g)", "Sugars, total\n (g)"]]
nutrition1 = nutrition1.replace(0.0, np.nan)
nutrition1["Protein(g)"] = nutrition1["# 100 grams"] * nutrition1["Protein (g)"]
nutrition1["Carbohydrates(g)"] = nutrition1["# 100 grams"] * nutrition1["Carbohydrate (g)"]
nutrition1["Fiber(g)"] = nutrition1["# 100 grams"] * nutrition1["Fiber, total dietary (g)"]
nutrition1["Fat(g)"] = nutrition1["# 100 grams"] * nutrition1["Total Fat (g)"]
nutrition1["Sugar(g)"] = nutrition1["# 100 grams"] * nutrition1["Sugars, total\n (g)"]
nutrition1.drop(columns =["Protein (g)","Carbohydrate (g)","Fiber, total dietary (g)","Total Fat (g)","Sugars, total\n (g)","# 100 grams"])
nutrition1
```

```python
nutrition2 = nutrition1.copy(deep = True)
nutrition2["Convenience Cost"] = WICcosts["7-11"]
nutrition2["Corner Cost"] = WICcosts["Family Dollar"]
nutrition2["Grocery Cost"] = WICcosts["America's Food Basket (local grocery store)"]
nutrition2["Superstore Cost"] = WICcosts["Stop and Shop"]
```

```python
nutrition3 = nutrition2.copy(deep = True)
nutrition3["Cost per g Protein - Superstore"] = nutrition2["Superstore Cost"]/nutrition2["Protein(g)"]
nutrition3["Cost per g Carbohydrates - Superstore"] = nutrition2["Superstore Cost"]/nutrition2["Carbohydrates(g)"]
nutrition3["Cost per g Fiber - Superstore"] = nutrition2["Superstore Cost"]/nutrition2["Fiber(g)"]
nutrition3["Cost per g Sugar - Superstore"] = nutrition2["Superstore Cost"]/nutrition2["Sugar(g)"]
nutrition3["Cost per g Fat - Superstore"] = nutrition2["Superstore Cost"]/nutrition2["Fat(g)"]

nutrition3["Cost per g Protein - Grocery"] = nutrition2["Grocery Cost"]/nutrition2["Protein(g)"]
nutrition3["Cost per g Carbohydrates - Grocery"] = nutrition2["Grocery Cost"]/nutrition2["Carbohydrates(g)"]
nutrition3["Cost per g Fiber - Grocery"] = nutrition2["Grocery Cost"]/nutrition2["Fiber(g)"]
nutrition3["Cost per g Sugar - Grocery"] = nutrition2["Grocery Cost"]/nutrition2["Sugar(g)"]
nutrition3["Cost per g Fat - Grocery"] = nutrition2["Grocery Cost"]/nutrition2["Fat(g)"]

nutrition3["Cost per g Protein - Corner"] = nutrition2["Corner Cost"]/nutrition2["Protein(g)"]
nutrition3["Cost per g Carbohydrates - Corner"] = nutrition2["Corner Cost"]/nutrition2["Carbohydrates(g)"]
nutrition3["Cost per g Fiber - Corner"] = nutrition2["Corner Cost"]/nutrition2["Fiber(g)"]
nutrition3["Cost per g Sugar - Corner"] = nutrition2["Corner Cost"]/nutrition2["Sugar(g)"]
nutrition3["Cost per g Fat - Corner"] = nutrition2["Corner Cost"]/nutrition2["Fat(g)"]

nutrition3["Cost per g Protein - Convenience"] = nutrition2["Convenience Cost"]/nutrition2["Protein(g)"]
nutrition3["Cost per g Carbohydrates - Convenience"] = nutrition2["Convenience Cost"]/nutrition2["Carbohydrates(g)"]
nutrition3["Cost per g Fiber - Convenience"] = nutrition2["Convenience Cost"]/nutrition2["Fiber(g)"]
nutrition3["Cost per g Sugar - Convenience"] = nutrition2["Convenience Cost"]/nutrition2["Sugar(g)"]
nutrition3["Cost per g Fat - Convenience"] = nutrition2["Convenience Cost"]/nutrition2["Fat(g)"]

nutrition3
```

```python
nutrition4 = nutrition3.copy(deep = True)
nutrition4 = nutrition4.drop(columns = ["# 100 grams", "Protein (g)", "Carbohydrate (g)", "Fiber, total dietary (g)", "Total Fat (g)", "Sugars, total\n (g)","Protein(g)","Carbohydrates(g)", "Fiber(g)", "Fat(g)", "Sugar(g)","Convenience Cost","Corner Cost", "Grocery Cost", "Superstore Cost", "Food and Product Sizes"])
nutrition4
```
